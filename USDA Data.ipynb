## Second Request

```python
API2 = "https://api.ers.usda.gov/data/arms/surveydata?api_key=AUaH6PHyXvFlp24BSQBnDlrFBhf2IwbS9SbDNmcZ"

years2 = [2011, 2012, 2013, 2014, 2015, 2016]


years_str2 = ','.join(map(str, years))

parameters2 = {
    "year": years_str2,  
    "state": 'all',
    "variable": 'Total acres operated',
    "category": 'Economic Class',
    "category2": 'Operator age',
    "report": 'Structural Characteristics',
}

response2 = requests.get(API2, params=parameters2)

if response2.status_code == 200:
    data2 = response2.json()
    
else:
    print(f"API request failed with status code {response2.status_code}")
```

## Joining tables

```python
Data_model = pd.concat([Dataset_1, Dataset_2], axis=0).sort_values('year')
Data_model.reset_index(drop=True, inplace=True)
Data_model
```

## Buckets

```python
bucket_filtered = Data_model[Data_model['variable_name'] == 'Total value of production']
```

```python
bucket_quartile = np.percentile(bucket_filtered['estimate'],[25,50,75])
bucket_quartile
```

```python
bins = [0,bucket_quartile[0],bucket_quartile[1],bucket_quartile[2],bucket_filtered['estimate'].max()]
bins
```

```python
Data_model['intervals_estimate'] = pd.cut(bucket_filtered['estimate'], bins)
Data_model
```

## Random Dates

```python
random_dates = pd.date_range(start="2000-01-01", end="2000-12-31", periods=365)
```

```python
Data_model['random_date'] = np.random.choice(random_dates, len(Data_model))

Data_model
```

```python
Data_model['random_date'] = pd.to_datetime(Data_model['random_date'])
Data_model['random_date'] = Data_model['random_date'].dt.date
```

```python
Data_model['random_date'] = pd.to_datetime(Data_model['random_date'])

def update_random_date(row):
    
        return row['random_date'].replace(year=row['year'])
   

Data_model['random_date'] = Data_model.apply(update_random_date, axis=1)

Data_model
```

## Julian date

```python
Data_model['random_date'] = pd.to_datetime(Data_model['random_date']).values.astype('datetime64[D]')

## Manual calculation
Data_model['julian_date'] = (Data_model['random_date'] - np.datetime64('1970-01-01')).dt.total_seconds() / 86400

```

```python
Data_model.to_csv('ARMS_test.csv')
```
